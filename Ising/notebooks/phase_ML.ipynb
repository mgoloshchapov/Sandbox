### Generating dataset

- Monte-Carlo single-spin flip dynamics is used to simulate Ising model


- Ferromagnetic nearest neighbour case is considered


- First we try to train CNN to distinguish between ferromagnetic and paromagnetic phases

```python
@njit
def ising_energy(field, J=1):
    n = field.shape[0]
    
    energy = 0
    for i in range(n):
        for j in range(n):
            energy += -J*field[i, j] * (field[i, (j+1)%n] + field[i, (j-1)%n] + field[(i+1)%n,j] + field[(i-1)%n,j])
    return energy



@njit
def ising_simulation(field, J=1, T=0.05, maxiter=int(1e8), step=int(1e6), eps=5e-4):
    #init
    n = field.shape[0]
    
    #to flip or not to flip
    rand_positions = np.random.randint(low=0, high=n, size=2*maxiter).reshape(-1, 2)
    rand_decision = np.random.random(maxiter)

    temp_E = ising_energy(field)
    #Ising steps, Monte-Carlo
    for idx in range(maxiter):
        i, j = rand_positions[idx]
        energy = -J*field[i, j] * (field[i, (j+1)%n] + field[i, (j-1)%n] + field[(i+1)%n,j] + field[(i-1)%n,j])

        if energy > 0 or rand_decision[idx] < np.exp(2*energy/T):
            field[i, j] = -field[i, j]
    
        if (idx+1) % step == 0:
            temp_E2 = ising_energy(field)
            delta = np.abs(temp_E2 - temp_E)
            
            if delta < eps * np.abs(temp_E):
                break
            else:
                temp_E = temp_E2
                
    return field, np.abs(delta/temp_E), idx+1
```

```python
n = 100
field = np.random.choice([-1,1], p=[0.5, 0.5], size=(n, n))
field = ising_simulation(field, J=1.0, T=3.0, maxiter=int(1e7))[0]

plt.figure(figsize=(10,10))
plt.imshow(field, cmap='magma', vmin=-1.0, vmax=1.0)
plt.xticks([])
plt.yticks([])
```

- temperatures from 0.25 to 3.00 with step 0.05, 1000 images for each temperature


- train/test split 80/20

```python
T = np.arange(0.25, 3.05, 0.05)
T = np.repeat(T, 1000)
fields = []
labels = []

for t in T:
    field = np.random.choice([-1,1], p=[0.5, 0.5], size=(n, n))
    field = ising_simulation(field, J=1.0, T=t, maxiter=int(2e6))[0]
    fields.append(copy.deepcopy(field))
    labels.append(round(t, 2))
    
fields = np.array(fields)

np.save("ising_data.npy", fields)
np.save("ising_labels.npy", labels)
```

### Phase transition prediction


- let's learn CNN to distinguish between ordered and disordered phases

```python
images = np.load("ising_data.npy")
labels = np.load("ising_labels.npy")
```

```python
images.shape[0]
```

```python
shuffled_idx = np.arange(images.shape[0])
np.random.shuffle(shuffled_idx)

train_idx, test_idx = shuffled_idx[:9*images.shape[0]//10], shuffled_idx[9*images.shape[0]//10:]
print(train_idx.shape)
print(test_idx.shape)
```

```python
plt.figure(figsize=(6,7))
for i in range(1,17,1):
    plt.subplot(4,4,i)
    plt.imshow(images[3000*i], cmap='magma')
    plt.xticks([])
    plt.yticks([])
    plt.title("T={}".format(labels[3000*i],fontsize=15))
plt.tight_layout()
# plt.savefig("ising_training.pdf", dpi=300)
```

__Split train/test in 90/10__

```python
labels_new = copy.deepcopy(labels)
bins, cuts = np.histogram(labels_new, bins=[0.0, 1.0, 1.5, 1.9, 2.2, 2.5, 3.0])

a = 0
i = 0
for b in bins:
    labels_new[a:a+b] = i
    i += 1
    a += b
labels_new 
```

```python
train_images, test_images = images[train_idx, ::, ::], images[test_idx, ::, ::]
train_labels, test_labels = labels_new[train_idx], labels_new[test_idx]
```

__Create Convolutional Neural Network__

```python
model = models.Sequential()

model.add(
    layers.AveragePooling2D(pool_size=(2,2),
                     input_shape=(100,100,1))
)
model.add(
    layers.Conv2D(
        filters=32,
        kernel_size=(3, 3),
        activation='relu',
        
    )
)
model.add(
    layers.MaxPool2D(pool_size=(2,2))
)
model.add(
    layers.Conv2D(
        filters=32,
        kernel_size=(3,3),
        activation='relu'
    )
)
model.add(
    layers.MaxPool2D(pool_size=(2,2))
)
model.add(
    layers.Conv2D(
        filters=32,
        kernel_size=(3,3),
        activation='relu'
    )
)
model.add(
    layers.MaxPool2D(pool_size=(2,2))
)
model.add(layers.Flatten())
model.add(layers.Dense(64))
model.add(layers.Dense(6))
```

```python
model.summary()
```

```python
model.compile(
    optimizer='adam',
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    metrics=['accuracy']
)

history = model.fit(train_images, train_labels, epochs=5,
         validation_data=(test_images, test_labels))
```

```python
model.evaluate(test_images, test_labels)
```

```python
predict_labels = np.argmax(model.predict(test_images), axis=1)
```

```python
cf_matrix = confusion_matrix(test_labels, predict_labels, labels=[0,1,2,3,4,5], normalize='true')
```

```python
def generate_labels(bins):
    res = []
    for i in range(len(bins)-1):
        res.append(str(bins[i])+"-"+str(bins[i+1]))
    return res
```

```python
vis_labels = generate_labels([0.0, 1.0, 1.4, 1.8, 2.2, 2.6, 3.0])
```

```python
def plot_resutls(cf_matrix, predict_labels, vis_labels):
    plt.figure(figsize=(7,6))
    res = heatmap(cf_matrix, annot=True, cmap='Blues', 
            xticklabels=vis_labels, 
            yticklabels=vis_labels,
           annot_kws={"fontsize":14},
                linewidths=1,
                 linecolor="black",
                 clip_on=False)
    res.set_xticklabels(res.get_xmajorticklabels(), fontsize = 14)
    res.set_yticklabels(res.get_xmajorticklabels(), fontsize = 14)

    cax = res.figure.axes[-1]
    cax.tick_params(labelsize=14)
    res.add_patch(Rectangle((4, 4), 1, 1, edgecolor='red', fill=False, lw=3))
    plt.title("Ising CNN", fontsize=18)

    plt.tight_layout()
```

```python
plot_resutls(cf_matrix, predict_labels, vis_labels)
plt.savefig("ising_confusion.jpg", dpi=300)
```
