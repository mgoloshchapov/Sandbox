```python
def gauss(x,mu,sigma,A):
    return A*np.exp(-(x-mu)**2/(2 * sigma**2))

def bimodal(x,mu1,sigma1,A1, mu2, sigma2, A2, k):
    return gauss(x,mu1,sigma1,A1)+gauss(x,mu2,sigma2,A2) + k
```

```python
def bivoigt(x, mu1, sigma1, gamma1, A1, mu2, sigma2, gamma2, A2, C):
    return A1 * voigt_profile(x - mu1, sigma1, gamma1) + A2 * voigt_profile(x - mu2, sigma2, gamma2) + C


def bivoigt_equal(x, mu1, mu2, sigma, gamma, A1, A2):
    return A1 * voigt_profile(x - mu1, sigma, gamma) + A2 * voigt_profile(x - mu2, sigma, gamma) + C


def bimodal(x,mu1,sigma1,A1, mu2, sigma2, A2, C):
    return gauss(x,mu1,sigma1,A1)+gauss(x,mu2,sigma2,A2) + C


def bimodal_equal(x,mu1, mu2, sigma, A1, A2):
    return gauss(x,mu1,sigma,A1) + gauss(x,mu2,sigma,A2)


def gauss_fit(x, y, expected, y_label=None):
    params = curve_fit(bimodal_equal, x, y, expected)[0]
    mu1, mu2, sigma = params[:3]
    coef = 228.2 / abs(mu1 - mu2)
    x_new = (x - mu1) * coef
    mu1, mu2 = 0, (mu2-mu1) * coef
    sigma *= coef

    params = (mu1, mu2, sigma, params[3], params[4])


    plt.plot(x_new, bimodal_equal(x_new, *params))

    plt.figure(figsize=(12, 6))
    plt.plot(x_new, y,
             label="data",
             color="blue",
             alpha=0.2)

    plt.plot(x_new, bimodal_equal(x_new, *params),
             label="$\\sigma={}$".format(round(sigma, 2)),
             color="red")

    plt.xlabel("MHz", fontsize=14)
    plt.ylabel(y_label, fontsize=14)
    plt.title("Fit by bimodal", fontsize=14)

    plt.legend(fontsize=14)
    plt.plot()

    return params, x_new


def fit(x, y, expected, y_label=None):
    """
    :param x:  independent variable
    :param y:  dependent variable
    :param expected: expected parameters
    :param v_coef:
    :return: optimal (mu1, mu2, sigma, gamma, A1, A2)
    """

    res = gauss_fit(x, y, expected)
    mu1, mu2, sigma, A1, A2 = res[0]
    x_new = res[1]

    gauss_expected = (mu1, mu2, sigma, -20, A1, A2)
    print(gauss_expected)

    params = curve_fit(bivoigt_equal, x_new, y, gauss_expected)[0]

    mu1, mu2, sigma, gamma = params[:4]
    coef = 228.2 / abs(mu1 - mu2)
    x_new = (x_new - mu1) * coef
    mu1, mu2 = 0, (mu2-mu1) * coef
    sigma *= coef
    gamma *= coef

    params = (mu1, mu2, sigma, gamma, A1, A2)

    plt.figure(figsize=(12, 6))
    plt.plot(x_new, y,
             label="data",
             color="blue",
             alpha=0.3)

    plt.plot(x_new, bivoigt_equal(x_new, *params),
             label="$\\sigma={}, \\gamma={}$".format(sigma, gamma),
             color="red")

    plt.xlabel("MHz", fontsize=14)
    plt.ylabel(y_label, fontsize=14)

    plt.legend(fontsize=14)
    plt.plot()
    return params
```

```python
def least_sqruares(y, y_pred):
    return np.linalg.norm(y_pred - y)

def loss_function(par, x, y, func=bivoigt_equal):
    y_pred = func(par, x)
    return least_sqruares(y, y_pred)

def bivoigt_equal(params, x):
    mu1, mu2, sigma, gamma, A1, A2 = params[0], params[1], params[2], params[3], params[4], params[5]
    return A1 * voigt_profile(x - mu1, sigma, gamma) + A2 * voigt_profile(x - mu2, sigma, gamma)




df16 = pd.read_csv("16.csv")

x16 = np.array(df16.iloc[::, 0])[::100]
y16 = np.array(df16.iloc[::, 1])[::100]
cut = len(x16) // 3
C = y16[:cut].mean()
y16 -= C
x16 = x16[cut:]
y16 = y16[cut:]
expected = (0.0586, 0.0591, 0.0002, 4.5, 5.5)
# params = fit(x16, y16, expected)

result = gauss_fit(x16, y16, expected)
mu1, mu2, sigma, A1, A2 = result[0]
x_new = result[1]
gauss_expected = (mu1, mu2, sigma, 0, A1, A2)
print(gauss_expected)

low = (-10, 218.2, 0, 0, 20, 150)
high = (10, 238.2, 100, 50, 200, 500)


res = dual_annealing(loss_function,
                     bounds=list(zip(low, high)),
                     x0=gauss_expected,
                    args=(x_new, y16),
                    seed=42,
                     initial_temp=10,
                     maxiter=1000,
                     visit=2,
                     accept=-1,
                    no_local_search=False,
                    minimizer_kwargs={'method': 'BFGS'})

print(res.x)
```
