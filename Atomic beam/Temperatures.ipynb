```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.optimize import curve_fit, dual_annealing
from scipy.special import voigt_profile
from scipy.signal import savgol_filter
from scipy.optimize import dual_annealing
```

```python
def gauss(x,mu,sigma,A):
    return A * np.exp(-(x-mu)**2 / (2 * sigma**2))


def bimodal(x,mu1,sigma1,A1, mu2, sigma2, A2, C):
    return gauss(x,mu1,sigma1,A1) + gauss(x,mu2,sigma2,A2)


def bimodal_equal(x, mu1, mu2, sigma, A1, A2):
    return A1 * np.exp(-(x-mu1)**2 / (2 * sigma**2)) + A2 * np.exp(-(x-mu2)**2 / (2 * sigma**2))


def bivoigt_equal(x, mu1, mu2, sigma, gamma, A1, A2):
    return A1 * voigt_profile(x - mu1, sigma, gamma) + A2 * voigt_profile(x - mu2, sigma, gamma)


def bivoigt_optimize(params, x):
    return bivoigt_equal(x, *params)


def gauss_fit(x, y, expected, y_label=None):
    params = curve_fit(bimodal_equal, x, y, expected)[0]
    mu1, mu2, sigma = params[:3]
    coef = 228.2 / abs(mu2 - mu1)

    x_new = (x - mu1) * coef
    mu1, mu2 = 0, (mu2-mu1) * coef
    sigma *= coef

    new_params = (mu1, mu2, sigma, params[3], params[4])

    plt.figure(figsize=(12, 6))
    plt.plot(x_new, y,
             label="data",
             color="blue",
             alpha=0.2)

    plt.plot(x_new, bimodal_equal(x_new, *new_params),
             label="$\\sigma={}$".format(round(sigma, 2)),
             color="red")

    plt.xlabel("MHz", fontsize=14)
    plt.ylabel(y_label, fontsize=14)
    plt.title("Fit by bimodal", fontsize=14)

    plt.legend(fontsize=14)
    plt.plot()

    return new_params, x_new

def add_gamma(expected, gamma):
    mu1, mu2, sigma, A1, A2 = expected
    return (mu1, mu2, sigma/2, gamma, A1, A2)


def voigt_fit(x, y, expected, y_label=None):
    params = curve_fit(bivoigt_equal, x, y, expected)[0]
    mu1, mu2, sigma, gamma = params[:4]
    coef = 228.2 / abs(mu2 - mu1)

    x_new = (x - mu1) * coef
    mu1, mu2 = 0, (mu2-mu1) * coef
    sigma *= coef
    gamma *= coef

    new_params = (mu1, mu2, sigma, gamma, params[4], params[5])

    plt.figure(figsize=(12, 6))
    plt.plot(x_new, y,
             label="data",
             color="blue",
             alpha=0.2)

    plt.plot(x_new, bivoigt_equal(x_new, *new_params),
             label="$\\sigma={}, \\gamma={}$".format(round(sigma, 2), round(gamma, 2)),
             color="red")

    plt.xlabel("MHz", fontsize=14)
    plt.ylabel(y_label, fontsize=14)
    plt.title("Fit by bimodal", fontsize=14)

    plt.legend(fontsize=14)
    plt.plot()
```

```python
data1 = pd.read_csv("360.CSV")
x1, y1 = np.array(data1.iloc[::, 0]), np.array(data1.iloc[::, 1])
plt.figure(figsize=(10,5))
plt.plot(x1, y1)

a1, b1 = int(len(x1)*0.46), int(len(x1)*0.53)
x1, y1 = x1[a1:b1], y1[a1:b1]
y1 -= y1[0]

plt.figure(figsize=(10,5))
plt.plot(x1, y1)

```

```python
expected1 = (0.00325, 0.0039, 0.00025, 2.5, 5.5)
params1, x_new1 = gauss_fit(x1, y1, expected1)
```

```python
# voigt_fit(x_new1, y1, add_gamma(expected1, 40.0))
def least_sqruares(y, y_pred):
    return np.linalg.norm(y_pred - y)

def loss_function(par, x, y, func=bivoigt_optimize):
    y_pred = func(par, x)
    return least_sqruares(y, y_pred)

low = (-10, 218.2, 0, 20, 100, 100)
high = (10, 238.2, 40, 40, 1000, 1000)

res = dual_annealing(loss_function,
                     bounds=list(zip(low, high)),
                     x0=add_gamma(params1, 30),
                    args=(x_new1, y1),
                    seed=42,
                     initial_temp=10,
                     maxiter=1000,
                     visit=2,
                     accept=-1,
                    no_local_search=False,
                    minimizer_kwargs={'method': 'BFGS'})

plt.plot(x_new1, bivoigt_equal(x_new1, *res.x))
plt.plot(x_new1, y1)
print(res.x)
```
